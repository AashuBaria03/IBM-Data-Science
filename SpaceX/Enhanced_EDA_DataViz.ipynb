# Enhanced EDA & Data Visualization Notebook

This notebook includes:
- Data Loading
- Missing Value Analysis
- Descriptive Statistics
- Correlation Matrix & Heatmap
- Boxplots for Outlier Detection
- Histograms & KDE for Distribution
- Pairplots for Feature Relationships


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load dataset
df = pd.read_csv('dataset_part_2.csv')
df.head()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Descriptive statistics
df.describe()
```

```python
# Select only numeric columns
numeric_df = df.select_dtypes(include=[np.number])

# Plot Correlation Heatmap
plt.figure(figsize=(12,8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

```

```python
# Boxplots to check outliers
df.plot(kind='box', subplots=True, layout=(3,4), figsize=(15,10), sharex=False, sharey=False)
plt.tight_layout()

plt.show()
```

```python
# Select only numeric columns
numeric_df = df.select_dtypes(include=[np.number])

# Histograms
numeric_df.hist(bins=30, figsize=(15,10), color='skyblue', edgecolor='black')
plt.tight_layout()
plt.show()

# KDE Plots
numeric_df.plot(kind='density', subplots=True, layout=(3,4), figsize=(15,10), sharex=False)
plt.tight_layout()
plt.show()

```
